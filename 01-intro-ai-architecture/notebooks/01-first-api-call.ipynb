# 01: First API Call with Gemini

In this notebook, you'll make your first API call to Google's Gemini model.

## Prerequisites

- Python 3.10+
- `openai` installed (`pip install openai`)
- A OpenAI API key from [OpenAI Platform](https://platform.openai.com/apikey)

## 1. Setup

```python
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY")
if not API_KEY:
    print("WARNING: OPENAI_API_KEY not found. Create a .env file with your key.")
else:
    print(f"API Key loaded: {API_KEY[:8]}...")
```

```python
import openai

API_KEY = os.getenv("OPENAI_API_KEY")
client = openai.OpenAI(api_key=API_KEY)

model = openai.OpenAI("gpt-4o-mini")
print(f"Model loaded: gpt-4o-mini")
```

## 2. Your First Prompt

```python
response = client.chat.completions.create("Explain what an AI Engineer does in 3 sentences.")
print(response.choices[0].message.content)
```

## 3. Understanding the Response

Let's inspect the response object to understand what we get back from the API.

```python
print("Response object type:", type(response))
print("\nFull response:", response)
print("\nCandidates:", response.candidates)
print("\nUsage metadata:", response.usage)
```

## 4. Controlling Output with Parameters

Key parameters:
- **temperature**: Controls randomness (0.0 = deterministic, 1.0 = creative)
- **max_output_tokens**: Limits response length
- **top_p**: Nucleus sampling (alternative to temperature)

```python
generation_config = (
    temperature=0.3,
    max_output_tokens=200,
)

model = openai.OpenAI(
    "gpt-4o-mini",
    generation_config=generation_config
)

response = client.chat.completions.create("List 5 real-world applications of AI in e-commerce.")
print(response.choices[0].message.content)
```

## 5. Chat Conversation

Gemini supports multi-turn conversations with memory.

```python
chat = client.chat.completions.create(history=[])

response = chat.send_message("Hi, I'm learning AI engineering.")
print("Bot:", response.choices[0].message.content)

response = chat.send_message("What should I learn first?")
print("Bot:", response.choices[0].message.content)

print("\nChat history:")
for msg in chat.history:
    print(f"  {msg.role}: {msg.parts[0].text[:80]}...")
```

## 6. Structured Output

You can ask the model to return structured responses.

```python
import json

prompt = """
Compare these two AI roles as JSON:
1. Data Scientist
2. AI System Engineer

Return a JSON array with objects containing: role, focus, skills (array), avg_salary_range.
"""

response = client.chat.completions.create(prompt)
print(response.choices[0].message.content)
```
