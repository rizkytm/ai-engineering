## 2. Scope Definition Best Practices

Scope creep kills freelance projects. Define boundaries early.

```python
scope_framework = {
    "IN SCOPE (be specific):": [
        "What you WILL build",
        "Number of revisions included",
        "What 'done' looks like (testable criteria)",
        "Support period after delivery",
        "Response time during support period",
    ],
    "OUT OF SCOPE (explicitly state):": [
        "What is NOT included",
        "Features that require separate quotes",
        "Maintenance beyond support period",
        "Training or documentation beyond standard",
        "Third-party costs (API fees, hosting)",
    ],
    "CHANGE PROCESS:": [
        "Any scope changes require written approval",
        "New requests get a change order with pricing",
        "Timeline adjusts accordingly",
        "Minimum change order: [your minimum]",
    ],
    "ACCEPTANCE CRITERIA:": [
        "List what the client will verify before final payment",
        "Each deliverable has a clear definition of done",
        "Testing period: X business days per deliverable",
        "Sign-off required before proceeding to next phase",
    ]
}

for section, items in scope_framework.items():
    print(f"\n{section}")
    for item in items:
        print(f"  • {item}")

print("\n" + "="*50)
print("Golden Rule: If it's not in the scope document, it's not in scope.")
```

## 4. Pricing Calculator

Use this framework to estimate project pricing.

```python
def calculate_fixed_price(
    hourly_rate,
    estimated_hours,
    complexity_multiplier=1.3,
    buffer_percent=0.25,
    profit_margin=0.2
):
    """Calculate fixed price with risk buffers."""
    
    base_cost = hourly_rate * estimated_hours
    complexity_adjusted = base_cost * complexity_multiplier
    with_buffer = complexity_adjusted * (1 + buffer_percent)
    final_price = with_buffer * (1 + profit_margin)
    
    print(f"Pricing Breakdown:")
    print(f"  Base cost: {hourly_rate}/hr × {estimated_hours} hrs = ${base_cost:,.0f}")
    print(f"  Complexity adjusted (×{complexity_multiplier}): ${complexity_adjusted:,.0f}")
    print(f"  With buffer (+{buffer_percent*100}%): ${with_buffer:,.0f}")
    print(f"  Final price (+{profit_margin*100}% margin): ${final_price:,.0f}")
    print(f"\n  Effective hourly rate: ${final_price / estimated_hours:,.0f}/hr")
    
    return final_price

# Example: RAG system for a startup
print("Example: RAG Knowledge Base for a Startup\n")
price = calculate_fixed_price(
    hourly_rate=150,
    estimated_hours=120,
    complexity_multiplier=1.3,
    buffer_percent=0.25,
    profit_margin=0.2
)
```

## 5. Common Pricing Pitfalls

```python
pitfalls = [
    {
        "pitfall": "Undercharging to win the job",
        "why_it_happens": "Fear of losing the client",
        "solution": "Set a minimum rate and never go below it. Cheap attracts bad clients.",
    },
    {
        "pitfall": "Not charging for discovery",
        "why_it_happens": "It's just a quick call",
        "solution": "Discovery is billable work. Charge $200-500 for a proper audit.",
    },
    {
        "pitfall": "Ignoring API/hosting costs",
        "why_it_happens": "Forgetting to factor in ongoing expenses",
        "solution": "Itemize third-party costs separately. Never absorb them.",
    },
    {
        "pitfall": "No change order process",
        "why_it_happens": "Wanting to be helpful",
        "solution": "Every change gets documented and priced. No exceptions.",
    },
    {
        "pitfall": "Paying yourself only for build time",
        "why_it_happens": "Forgetting meetings, emails, research",
        "solution": "Include all time in your estimates, not just coding.",
    },
    {
        "pitfall": "Not raising rates over time",
        "why_it_happens": "Comfort with current rate",
        "solution": "Raise rates 10-20% every 6 months. New clients only.",
    },
]

for i, p in enumerate(pitfalls, 1):
    print(f"{i}. {p['pitfall']}")
   print(f"   Why: {p['why_it_happens']}")
   print(f"   Fix: {p['solution']}")
   print()
```

## 7. Proposal Red Flags (Run Away)

```python
red_flags = {
    "Budget": [
        """We don't have a budget in mind""" → They're not serious or can't afford you,
        """Can you do it for equity?""" → Rarely worth it unless YC-backed,
        """We found someone cheaper""" → Let them go, don't race to the bottom,
    ],
    "Scope": [
        """Just build what [big company] has""" → Unrealistic expectations,
        """We'll figure it out as we go""" → Scope creep guaranteed,
        """How hard can it be?""" → They don't respect the work,
    ],
    "Communication": [
        """We need it yesterday""" → Poor planning on their part,
        """Can you start before signing?""" → Never work without a contract,
        """We'll pay you after we see results""" → No. Get a deposit,
    ],
    "Control": [
        """We need to approve every line of code""" → Micromanagement ahead,
        """Our intern will manage you""" → Wrong power dynamic,
        """We've tried 3 developers already""" → They're the problem,
    ]
}

print("RED FLAGS IN CLIENTS — WALK AWAY IF YOU SEE THESE:\n")

for category, flags in red_flags.items():
    print(f"{category}:")
    for flag in flags:
        print(f"  🚩 {flag}")
    print()
```

## Summary

1. **Use the proposal template** for every project — customize the details

3. **Choose the right pricing model** based on project clarity
4. **Always include buffers** — 20-30% for unknowns
5. **Send regular updates** — silence kills trust
6. **Walk away from red flags** — not every client is worth having
