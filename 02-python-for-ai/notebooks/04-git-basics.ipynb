## 1. What is Git?

Git is a **distributed version control system** that tracks changes to your code over time.

### Why Git Matters for AI Engineers:
- **Track experiments**: Know which code produced which results
- **Collaborate**: Work with teammates on the same codebase
- **Revert mistakes**: Undo changes that broke something
- **Branch safely**: Try new ideas without breaking the main code
- **Reproduce results**: Go back to the exact code used for a model

### Key Concepts

| Concept | What it is |
|---------|------------|
| **Repository (repo)** | A project folder tracked by Git |
| **Commit** | A snapshot of your code at a point in time |
| **Branch** | An independent line of development |
| **Merge** | Combining changes from different branches |
| **Remote** | The version of your repo hosted on GitHub |
| **Clone** | Copying a remote repo to your local machine |
| **Staging** | Preparing changes to be included in the next commit |

## 3. Core Git Commands

### 3.1 Creating a Repository

```bash
# Initialize a new repo in the current directory
git init

# Clone an existing repo from GitHub
git clone https://github.com/username/repo-name.git
```

### 3.2 The Git Workflow

The basic cycle is: **Edit -> Stage -> Commit**

```
Working Directory  --git add-->  Staging Area  --git commit-->  Repository
     (edit files)              (prepare snapshot)             (save snapshot)
```

### 3.3 Checking Status

```bash
# See what's changed
git status

# Output example:
# On branch main
# Changes not staged for commit:
#   modified:   train.py
# Untracked files:
#   results.json
```

### 3.4 Staging Changes

```bash
# Stage a specific file
git add train.py

# Stage multiple files
git add train.py utils.py

# Stage all changes
git add .

# Stage changes in a specific directory
git add src/
```

### 3.5 Committing

```bash
# Commit with a message
git commit -m "Add data preprocessing pipeline"

# Commit with a detailed message
git commit -m "Add data preprocessing pipeline" -m "
- Load CSV files with Pandas
- Handle missing values
- Normalize numeric columns
- Split into train/test sets"
```

### 3.6 Viewing History

```bash
# View commit history
git log

# Compact view
git log --oneline

# With file changes
git log --stat

# Graph view (shows branches)
git log --oneline --graph --all

# See what changed in a specific commit
git show abc1234

# See differences (unstaged)
git diff

# See differences (staged)
git diff --staged
```

### 3.7 Undoing Changes

```bash
# Discard changes in working directory (unstaged only)
git checkout -- file.py
# or (newer)
git restore file.py

# Unstage a file (keep changes)
git reset HEAD file.py
# or (newer)
git restore --staged file.py

# Amend the last commit (before pushing)
git commit --amend -m "Updated commit message"

# Revert a commit (creates a new commit that undoes changes)
git revert abc1234
```

## 4. Branches

Branches let you work on features or experiments without affecting the main codebase.

### 4.1 Branch Commands

```bash
# List branches
git branch          # Local branches
git branch -a       # All branches (local + remote)

# Create a new branch
git branch feature/new-model

# Switch to a branch
git checkout feature/new-model
# or (newer)
git switch feature/new-model

# Create and switch in one command
git checkout -b feature/new-model
# or
git switch -c feature/new-model

# Delete a branch
git branch -d feature/new-model   # Safe delete
git branch -D feature/new-model   # Force delete
```

### 4.2 Merging

```bash
# Merge a branch into the current branch
git checkout main
git merge feature/new-model

# If there are merge conflicts, you need to:
# 1. Edit the conflicted files
# 2. Remove conflict markers (<<<<<<<, =======, >>>>>>>)
# 3. Stage the resolved files
# 4. Commit
git add resolved-file.py
git commit
```

### Merge Conflict Example

```
<<<<<<< HEAD
learning_rate = 0.001
=======
learning_rate = 0.01
>>>>>>> feature/new-model
```

Choose the correct value, remove the markers, and commit.

## 5. Working with GitHub

GitHub is a hosting service for Git repositories. It adds collaboration features like pull requests, issues, and code review.

### 5.1 Remote Repositories

```bash
# Add a remote
git remote add origin https://github.com/username/repo.git

# List remotes
git remote -v

# Push to remote
git push origin main

# Push a new branch
git push -u origin feature/new-model

# Pull latest changes
git pull origin main

# Fetch (download changes without merging)
git fetch origin
```

### 5.2 The Pull Request Workflow

This is the standard way to contribute code on GitHub:

1. **Clone** the repository
   ```bash
   git clone https://github.com/username/repo.git
   ```

2. **Create** a feature branch
   ```bash
   git checkout -b feature/my-new-feature
   ```

3. **Make changes** and commit
   ```bash
   git add .
   git commit -m "Add new feature"
   ```

4. **Push** to GitHub
   ```bash
   git push -u origin feature/my-new-feature
   ```

5. **Open a Pull Request** on GitHub
   - Go to the repository on GitHub
   - Click "Compare & pull request"
   - Add a title and description
   - Request reviewers

6. **Review** and merge
   - Address review feedback
   - Merge when approved

## 7. Commit Best Practices

### Writing Good Commit Messages

```
Short summary (50 chars or less)

Optional longer explanation wrapped at 72 characters.
Explain the problem this commit solves, not how.

Fixes #42
```

### Rules:

1. **Use imperative mood**: "Add feature" not "Added feature" or "Adds feature"
2. **Keep the subject line under 50 characters**
3. **Start with a capital letter**
4. **Don't end the subject line with a period**
5. **Use the body to explain what and why, not how**
6. **Reference issues** when relevant ("Fixes #42", "Closes #15")

### Good vs Bad Commits

| Bad | Good |
|-----|------|
| `fixed stuff` | `Fix data loading bug in preprocessing pipeline` |
| `WIP` | `Add initial model training script with hyperparameter tuning` |
| `update` | `Update README with installation instructions` |
| `changes` | `Refactor API client to use connection pooling` |

## 8. Common Git Scenarios for AI Engineers

### Tracking Experiments

```bash
# Commit your experiment code with the results
git add train.py config.yaml results/metrics.json
git commit -m "Add experiment v3: fine-tune with lr=0.001, batch=32"
```

### Starting a New Project

```bash
# Create a new directory and initialize
mkdir my-ai-project
cd my-ai-project
git init

# Create initial files
touch README.md .gitignore requirements.txt
git add .
git commit -m "Initial commit: project structure"

# Connect to GitHub
git remote add origin https://github.com/username/my-ai-project.git
git push -u origin main
```

### Working on a Team

```bash
# Always pull before starting work
git pull origin main

# Create a branch for your task
git checkout -b feature/data-pipeline

# Do your work, commit, push
git add .
git commit -m "Add data pipeline with CSV loading"
git push -u origin feature/data-pipeline

# Open a PR on GitHub, get review, merge
```

## Summary

Key takeaways:

2. **Branches enable experimentation** — try ideas without breaking main
3. **Pull requests enable collaboration** — review code before merging
4. **Write good commit messages** — future you will thank present you
5. **Use .gitignore** — don't commit data files, secrets, or cache
