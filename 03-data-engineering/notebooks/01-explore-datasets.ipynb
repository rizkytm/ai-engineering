## 4. Exploring Data with Pandas

### 4.1 First Look: `head()`, `tail()`, `shape`

```python
# First 5 rows
print("=== Head ===")
df.head()

# Last 3 rows
print("\n=== Tail ===")
df.tail(3)
```

```python
# Shape: (rows, columns)
print(f"Shape: {df.shape}")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

### 4.2 Data Types and Structure: `info()`

```python
# Overview of the DataFrame
df.info()
```

**What to look for in `info()`:**
- Non-null counts (are there missing values?)
- Data types (are dates stored as objects instead of datetime?)
- Memory usage

### 4.3 Statistics: `describe()`

```python
# Statistics for numeric columns
df.describe()
```

```python
# Statistics for all columns including text
df.describe(include="all")
```

**Key stats to check:**
- `count`: Are all columns the same count? If not, some have missing values
- `min`/`max`: Any suspicious values? (e.g., negative age)
- `mean`/`std`: Is the data skewed?

### 4.4 Unique Values: `value_counts()`

```python
# Count unique values in categorical columns
print("=== Department Distribution ===")
print(df["department"].value_counts())

print("\n=== Performance Distribution ===")
print(df["performance"].value_counts())
```

```python
# Percentage distribution
print("=== Department % ===")
print(df["department"].value_counts(normalize=True).round(3) * 100)
```

### 4.5 Column Selection and Filtering

```python
# Select a single column
df["name"]
```

```python
# Select multiple columns
df[["name", "salary"]]
```

```python
# Filter rows: engineers with salary > 80000
df[(df["department"] == "Engineering") & (df["salary"] > 80000)]
```

### 4.6 Checking for Issues

```python
# Missing values per column
print("=== Missing Values ===")
print(df.isnull().sum())

# Total missing
print(f"\nTotal missing: {df.isnull().sum().sum()}")
```

```python
# Duplicates
print(f"Duplicate rows: {df.duplicated().sum()}")

# Check for exact duplicates
df[df.duplicated(keep=False)]
```

```python
# Column data types
print("=== Data Types ===")
print(df.dtypes)

# Check unique values per column
print("\n=== Unique Counts ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

## 6. Understanding Data Schemas

A data schema defines the structure, types, and constraints of your data.

```python
# Schema summary function
def describe_schema(df):
    """Generate a schema summary for a DataFrame."""
    schema = pd.DataFrame({
        "dtype": df.dtypes,
        "non_null": df.count(),
        "null_count": df.isnull().sum(),
        "null_pct": (df.isnull().sum() / len(df) * 100).round(1),
        "unique": df.nunique(),
        "sample": [df[col].dropna().iloc[0] if not df[col].dropna().empty else "N/A" for col in df.columns]
    })
    return schema

describe_schema(df)
```
