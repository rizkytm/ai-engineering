# 02: AI Architecture Diagrams

Practice drawing and analyzing AI system architectures.

## Why This Matters

Before building any AI system, you need to understand:
- What data flows through the system
- Which components interact
- Where the AI model sits in the stack

## 1. Architecture Layers Framework

Every AI system can be decomposed into these layers:

```
┌─────────────────────────────────┐
│       APPLICATION LAYER         │
│   (UI, API, Dashboard)          │
├─────────────────────────────────┤
│      ORCHESTRATION LAYER        │
│   (LangChain, Prompt, Tools)    │
├─────────────────────────────────┤
│       INFERENCE LAYER            │
│   (LLM calls, Embeddings)       │
├─────────────────────────────────┤
│         MODEL LAYER             │
│   (GPT-4, Gemini, Llama)        │
├─────────────────────────────────┤
│         DATA LAYER              │
│   (Vector DB, SQL, Files)       │
├─────────────────────────────────┤
│       DATA SOURCE LAYER         │
│   (Docs, APIs, User input)      │
└─────────────────────────────────┘
```

## 2. Case Study: AI Chatbot

Map each component of a customer support chatbot to the layers.

````python
chatbot_architecture = {
    "Application Layer": {
        "component": "Web Chat Widget",
        "description": "React component embedded in support page",
        "technology": "React, WebSocket"
    },
    "Orchestration Layer": {
        "component": "LangChain Chain",
        "description": "Manages prompt assembly, retrieval, and response generation",
        "technology": "LangChain, Prompt Templates"
    },
    "Inference Layer": {
        "component": "Gemini API",
        "description": "Text generation and embedding calls",
        "technology": "Google AI SDK"
    },
    "Model Layer": {
        "component": "GPT-4o Mini",
        "description": "Foundation model for generation",
        "technology": "Google Gemini"
    },
    "Data Layer": {
        "component": "Pinecone + PostgreSQL",
        "description": "Vector store for docs, relational DB for user data",
        "technology": "Pinecone, PostgreSQL"
    },
    "Data Source Layer": {
        "component": "Knowledge Base",
        "description": "FAQ docs, product manuals, past support tickets",
        "technology": "PDF, Markdown, API"
    }
}

for layer, details in chatbot_architecture.items():
    print(f"\n{'='*50}")
    print(f"{layer}")
    print(f"  Component: {details['component']}")
    print(f"  Purpose:   {details['description']}")
    print(f"  Tech:      {details['technology']}")
````

## 3. Mermaid Diagrams

Mermaid is a markdown-based diagramming tool. Many docs and wikis support it.

Here's how to represent an AI chatbot architecture:

````python
mermaid_diagram = """
graph TD
    A[User Input] --> B[Web Chat Widget]
    B --> C[FastAPI Backend]
    C --> D[LangChain Orchestrator]
    D --> E[Embedding - Gemini]
    D --> F[Vector Search - Pinecone]
    D --> G[LLM - Gemini Pro]
    F --> H[Knowledge Base]
    G --> I[Response]
    I --> B
    
    style A fill:#e1f5fe
    style G fill:#f3e5f5
    style F fill:#e8f5e9
"""

print("Mermaid syntax for AI Chatbot:")
print(mermaid_diagram)
print("\nPaste this into any Mermaid renderer (mermaid.live) to see the diagram.")
````

## 4. Data Flow Analysis

Understanding what data moves where is critical for:
- **Security**: What data goes to external APIs?
- **Latency**: Which paths are slow?
- **Cost**: Where are tokens being consumed?

````python
data_flow = [
    {
        "step": 1,
        "action": "User sends question",
        "data": "Raw text string",
        "cost": "Free",
        "latency": "~50ms"
    },
    {
        "step": 2,
        "action": "Generate embedding",
        "data": "Text -> 768-dim vector",
        "cost": "~$0.00001 per embedding",
        "latency": "~100ms"
    },
    {
        "step": 3,
        "action": "Vector search in Pinecone",
        "data": "Query vector -> top-k documents",
        "cost": "Free tier available",
        "latency": "~50ms"
    },
    {
        "step": 4,
        "action": "Assemble prompt",
        "data": "Context + question -> full prompt",
        "cost": "Free",
        "latency": "~10ms"
    },
    {
        "step": 5,
        "action": "Call Gemini Pro",
        "data": "Prompt -> response text",
        "cost": "~$0.001 per request",
        "latency": "~1-3s"
    },
    {
        "step": 6,
        "action": "Return to user",
        "data": "Formatted response",
        "cost": "Free",
        "latency": "~50ms"
    }
]

print(f"{'Step':<5} {'Action':<25} {'Data':<30} {'Cost':<25} {'Latency'}")
print("-" * 120)
for step in data_flow:
    print(f"{step['step']:<5} {step['action']:<25} {step['data']:<30} {step['cost']:<25} {step['latency']}")
````
